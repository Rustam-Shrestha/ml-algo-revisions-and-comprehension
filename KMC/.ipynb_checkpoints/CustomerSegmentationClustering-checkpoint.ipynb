```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('3.12. Example.csv')
```

```python
data
# this data is from retail shhop for brand loyalty
```

```python
plt.scatter(data['Satisfaction'], data['Loyalty'])
plt.xlabel("Satisfaction")
plt.ylabel('Loyalty')
plt.show()
```

# breaking this graph we may need to depict this graph as 


```python
x = data.copy()
```

# Clustering the data

```python
k_means = KMeans(2)
```

```python
k_means.fit(x)
```

# Clustering results

```python
clusters = x.copy()
clusters['predicted']=k_means.fit_predict(x)
plt.scatter(clusters['Satisfaction'], clusters['Loyalty'],c=clusters['predicted'], cmap='rainbow')
plt.xlabel("Satisfaction")
plt.ylabel('Loyalty')
plt.show()
```

# if we notice K means clustered data we see it considered the satisfaction as the cluster point because loyalty had lesser valeu and due to no standardization this just classified on the basis of satisfaction not on the basis of the important concept might be an from both and  disregarded by loyalty as feature

# Standardization fo the valuees 

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
```

```python
x_scaled
```

# if we see closely loyalty value seems like noot being scaled but in real it is already scaled so algorithm didnt scale it and only scaled satisfaction

# Elbow method

```python
wcss = []
for i in range (1,10):
    k_means = KMeans(i)
    k_means.fit(x)
    wcss.append(k_means.inertia_)    
```

```python
wcss
```

```python
plt.plot(range(1,10),wcss)
plt.xlabel("number of clusters")
plt.ylabel("WCSS")
```

# exploring cluster solutions and selectin  numbers of clusters
# tip or edge on the 2 3 4 5 gives us the optimal point where each indicates there is diminishing  returns

```python
# appending the standardized prediction in non standardized dataset to plot unstandardized data but standardized clustering
# Fiddle with K (the number of clusters)
k_means_new = KMeans(4)
# Fit the data
k_means_new.fit(x_scaled)
# Create a new data frame with the predicted clusters
clusters_new = x.copy()
clusters_new['cluster_pred_standardized'] = k_means_new.fit_predict(x_scaled)
```

```python
clusters_new
```

```python
# Plot
plt.scatter(clusters_new['Satisfaction'],clusters_new['Loyalty'],c=clusters_new['cluster_pred_standardized'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```


# now we see same type of segmentation using k means clustering
