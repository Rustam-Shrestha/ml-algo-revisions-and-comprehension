```python
import statsmodels.api as sn
import statsmodels.discrete.discrete_model as Logmod
from sklearn.linear_model import LogisticRegression
```

```python
y=df['Sold']
```

```python
clf_mlr = LogisticRegression()
clf_mlr.fit(x, y)
```

```python
x_const = sn.add_constant(x)
```
