```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sma
from sklearn.linear_model import LinearRegression

# Load the dataset
df = pd.read_csv("House_Price.csv")
```

```python

# Using a multiple regression using statmodels api
# we are dropping the column so we write axis=1 while dropping the row
#we write axis = 0 and instead of price we write index of row
x_multi = df.drop("price",axis=1)
x_multi.head()
```

```python

# creating the dependant variable
y_multi = df["price"]
y_multi.head()
```

```python
# adding the constant in the equation for beta which is by default 0 to 1
# basically y-mx+c so c will get 1 as first column printing x_multi_constant is 1
x_multi_konst = sma.add_constant(x_multi)
x_multi_konst.head()
```

```python
lm_multi = sma.OLS(y_multi, x_multi_konst).fit()
```
