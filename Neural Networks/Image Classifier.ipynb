### Split the data into train/validation/test datasets
In the earlier step of importing the date, we had 60,000 datasets for training and 10,000 test datasets. Now we further split the training data into train/validation. Here is how each type of dateset is used in deep learning:

* __Training data__ — used for training the model
* __Validation data__ — used for tuning the hyperparameters and evaluate the models
* __Test data__ — used to test the model after the model has gone through initial vetting by the validation set.

```python
# splitting data into test validation 
# x valid gets 5000 samples from x train normalized and x train gets rest of the data from normalzied data
X_valid, X_train = X_train_normalized[:5000], X_train_normalized[5000:]
# same for y_valid and train
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
#  new label for x_trest 
X_test = X_test_normalized
```

# now essential thing is designing model 
# there are two essential ways building models in keras
   * __Sequential model api__
   * __functional model api__

The Sequential API is the simplest way to build a model in Keras. It allows you to create models layer-by-layer in a linear stack. This is ideal for most straightforward problems where you have a single input and output.

The Functional API is more flexible and powerful. It allows you to build complex models, such as multi-input, multi-output models, models with shared layers, and models with non-linear topology (e.g., residual connections).

```python
# setting random values for initializing neural network
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# initializing sequential model wth linear stacked layer
img_cls_mdl = keras.models.Sequential()
# input the data with matrix of 28x28 image
img_cls_mdl.add(keras.Input(shape=[28, 28]))
# putting 28x28 into 768x1 matrix
img_cls_mdl.add(keras.layers.Flatten())
# Relu with 300 different neurons fully connected
img_cls_mdl.add(keras.layers.Dense(300, activation="relu"))
# next collection of relu to process the utput of the previous 300 relu neural perceptrons
img_cls_mdl.add(keras.layers.Dense(100, activation="relu"))
#  PRBABILITY DISTRIBUTION OVER THEN DIFFERENT CATEGORY WITH NEW LAYER WITH SOFTMAX ACTIVATION FUNCTION
img_cls_mdl.add(keras.layers.Dense(10, activation="softmax"))
```

__Softmax__: The softmax function takes a vector of raw scores (logits) from the previous layer and converts them into probabilities. Each value in the output vector is between 0 and 1, and the sum of all values is 1.

```python
# we are not dealing wigh regression we are dealing with categorical variables like sneakers, bag, trousers and more
# so the loss function is used as categorical x entropy 
# using stochastinc gradient descent optimization 
img_cls_mdl.compile(loss="sparse_categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])
```

```python
# fitting the mdoel as 
mdl_fit = img_cls_mdl.fit(X_train, y_train, epochs=30,
                    validation_data=(X_valid, y_valid))
```
