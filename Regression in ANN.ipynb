```python
from sklearn.model_selection import train_test_split
# splitting the dagta for training and testing
```

```python
housing.target
```

```python
# splitting the data with first paameter of array house tuples, target being teh median is 4.526 is 452,600
x_train, x_test, y_train, y_test = train_test_split(housing.data, housing.target, random_state = 4)
```

```python
# further splitting into validation
# splitting the data to 75/25 and further splitting 75 % into 75 25
x_train_next_split, x_valid, y_train_next_split, y_valid = train_test_split(x_train, y_train, random_state=42)
```

```python
# scaling the asymmetrical nature of data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
# reassigintng the scaled data rto the train test validasi
x_train = scaler.fit_transform(x_train)
x_valid = scaler.transform(x_valid)
X_test = scaler.transform(x_test)
```

```python
# Setting the random seed for reproducibility jmeaning someone reusing this with same randomn number get same output
np.random.seed(42)
tf.random.set_seed(42)
```

```python
x_train.shape
```

```python

# Sequentially training the model with 30 hidden perceptrons in each layer
trained_reg_ann_model = keras.models.Sequential([
    keras.layers.Input(shape=[8]),
    keras.layers.Dense(30, activation="relu"),
    keras.layers.Dense(30, activation="relu"),
    keras.layers.Dense(1)
])

# keras.layers.Dense(30, activation="relu", input_shape=X_train.shape[1:])
```

```python
trained_reg_ann_model.summary()
# 
```

# structure of the ANN now


### parameters taken by compile()
__loss:__ Specifies the loss function to be used. In your case, "mean_squared_error" is used for regression tasks.

__optimizer:__ Specifies the optimizer to be used for training. You are using keras.optimizers.SGD, which is the Stochastic Gradient Descent optimizer with a learning rate of 0.001.

__metrics:__ A list of metrics to be evaluated by the model during training and testing. You have specified mae, which stands for Mean Absolute Error.

__loss_weights:__ Optional list or dictionary specifying scalar coefficients to weight the loss contributions of different model outputs.

__weighted_metrics:__ List of metrics to be evaluated and weighted by sample_weight or class_weight during training and testing.

__run_eagerly:__ Boolean. If True, the model will run eagerly. This can be useful for debugging.

__steps_per_execution:__ Integer. The number of batches to run during each tf.function call. Running multiple batches inside a single tf.function call can greatly improve performance on TPUs or small models with a large Python overhead

```python
trained_reg_ann_model.compile(
    loss="mean_squared_error", 
    # using optimizer of stocastic gradient deswcent wit hlearning rate of 0.0001
    # This sets the learning rate for the optimizer to 0.001. The learning rate 
    # controls how much to change the model in response to the estimated error each time the model weights are updated.
    optimizer=keras.optimizers.SGD(learning_rate=1e-3), 
    metrics=['mae']
)
```

```python
# training the model with 20 iterative passing of data and 
model_information = trained_reg_ann_model.fit(x_train, y_train, epochs=20, validation_data=(x_valid, y_valid))
```

```python
mae_test = trained_reg_ann_model.evaluate(x_test, y_test)
```
