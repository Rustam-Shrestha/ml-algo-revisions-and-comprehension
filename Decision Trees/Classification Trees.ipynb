# this here we will classify movies prone to win an oscar or not which is categorica lvariable classificatioon

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('movie_classification.csv', header = 0)
```

```python
df.head()
```

```python
# drop n-1 will be faciliated by drop_first that will drop first of the extra vattribute generated
df = pd.get_dummies(df,columns = ['3D_available', 'Genre'], drop_first= True)
# Identify the dummy columns
dummy_columns = df.select_dtypes(include='bool').columns

# Convert boolean values to integers
df[dummy_columns] = df[dummy_columns].astype(int)

df.info()
```

```python
# Pandas 3.0 has this thing that will work as it takes json like object of atribute and replaces the na values
df.fillna({'Time_taken': df['Time_taken'].mean()}, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

# x and  y variable splitting of the dataset

```python
# locating all rows but excluding the main dpeendant variable
x =df.loc[:,  df.columns!="Start_Tech_Oscar"]
```

```python
x.head()
```

```python
y=df['Start_Tech_Oscar']
```

# splitting model for test and train 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=4)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn import tree
```

# classification Tree training

```python
clftree = tree.DecisionTreeClassifier(max_depth=3)
```

```python
clftree.fit(x_train, y_train)
```

```python
y_train_pred_clftree= clftree.predict(x_train)
```

```python
y_test_pred_clftree= clftree.predict(x_test)
```

```python
y_test_pred_clftree
```

# gauging the model performance

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_train, y_train_pred_clftree)
```

```python
confusion_matrix(y_test,y_test_pred_clftree)
```

```python
accuracy_score(y_test, y_test_pred_clftree)
```

# plotting the tree in pictoral format

```python
# creating dot file for plotting the regreession tree for viaualizing the filtering criteria for our tree
dot_data = tree.export_graphviz(clftree,feature_names=x_train.columns, out_file=None, filled=True)
# library to display the tree visualization
from IPython.display import Image
# we havbe to install this package to use it i first use pip install pydotplus in command procmpt
import pydotplus
# cinally creating and displaying 
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
# a decision tee with mazimum height o0f 3 nodes
```

# Bagging techniwue

```python
from sklearn import tree
clftree = tree.DecisionTreeClassifier()
```

```python
from sklearn.ensemble import BaggingClassifier as BClf
```

```python
bag_clf =BClf(clftree, n_estimators= 1000, bootstrap = True, n_jobs =-1, random_state=42)
```

# Hyperparamter tuning using sklearn and gridsearch library

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_clf_with_hyperparamter = RandomForestClassifier(n_estimators=250,random_state=42)
```

```python
grid_of_parameters = {'max_features': [4, 5, 6, 7, 8, 9, 10], 'min_samples_split': [2, 3, 10]}
```

```python
grid_search = GridSearchCV(rf_clf_with_hyperparamter, grid_of_parameters,n_jobs = -1,cv=5, scoring='accuracy')
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
random_forest_optimized = grid_search.best_estimator_
```

```python
accuracy_score(y_test, random_forest_optimized.predict(x_test))
```

# the best estimation can be 62.7% with only 250 samples and the hyperparamteer given so more sample will cause better
# performance which was 63 in 1000 samples so this way i did the grid search hyperparameter fixing
