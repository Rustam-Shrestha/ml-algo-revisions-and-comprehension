# doing the train-test split and evaluating it in logistic regression
# classifying algorithm for binary sold and not sold

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("D:\\Dashboard\\miDirectarios\\ML\\start tech python\\ML Algorithms\\Logistic Regression\\Classification preprocessed data Python.csv", header=0)
```

```python
df.info()
```

```python
# in sklearn our independant variablsem must be two dimensional so we use two brackets
x = df.drop(columns=['Sold'])
```

```python
y=df['Sold']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# due to the model not converging we set iteration to 900
clf_lr = LogisticRegression(max_iter=900)
```

```python
clf_lr.fit(x_train,y_train)
```

```python
y_test_predicted= clf_lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
confusion_matrix(y_test,y_test_predicted)
```

```python
accuracy_score(y_test,y_test_predicted)
```
